### PROJET TRAITEMENT DE DONNÉES ###

Dans un premier temps on importe la base de données 

```python
import pandas as pd
df = pd.read_csv(
 'https://www.data.gouv.fr/api/1/datasets/r/5bd2b320-fc25-4cce-a68d-bebd21ab0e85' 
)
df
```

Combien il a t'il de faux positifs (Mesure < Limite)? 


```python
df['faux_positifs'] = df['mesure'].lt(df['limite'], axis=0)

print(df)

faux_positifs = df['faux_positifs'].sum()
vrais_positifs = df['faux_positifs'].value_counts()[False]
print(f"Nombre de faux positifs : {faux_positifs}")
print(f"Nombre de vrais positifs : {vrais_positifs}")
```
